```python
import pandas as pd
import numpy as np
```

```python
covid = pd.read_csv('daily-covid-cases-deaths.csv')
covid = covid[covid['Entity'] == 'India']
covid.reset_index(inplace=True)
covid.drop(columns=['index'], inplace=True)
covid['Disease'] = 'Covid-19'
covid.rename(columns={'Daily new confirmed cases of COVID-19': 'Cases', 'Daily new confirmed deaths due to COVID-19': 'Deaths'}, inplace=True)
covid.to_csv('Daily/daily_covid-19.csv')
covid
```

```python
covid['Day'] = pd.to_datetime(covid['Day'])
covid['Year'] = covid['Day'].dt.year
covid = covid.groupby('Year')[['Cases', 'Deaths']].mean().reset_index()
covid['Entity'] = 'India'
covid['Code'] = 'IND'
covid['Disease'] = 'Covid-19'
covid.to_csv('Yearly/yearly_covid-19.csv')
covid
```

```python
influenza = pd.read_csv('influenza.csv')
influenza = influenza[influenza['Country'] == 'India']
influenza.reset_index(inplace=True)
influenza.drop(columns=['index'], inplace=True)
influenza_sum = influenza.iloc[:, 3:].sum(axis=1)
influenza = pd.concat([influenza.iloc[:, :2], influenza_sum], axis=1)
influenza['Code'] = 'IND'
influenza['Disease'] = 'Influenza'
influenza['Deaths'] = np.nan
influenza.rename(columns={'Country': 'Entity', 0: 'Cases', 'Month': 'Day'}, inplace=True)
influenza = influenza[['Entity', 'Code', 'Day', 'Cases', 'Deaths', 'Disease']]
influenza.to_csv('Daily/daily_influenza.csv')
influenza
```

```python
influenza['Day'] = pd.to_datetime(influenza['Day'])
influenza['Year'] = influenza['Day'].dt.year
influenza = influenza.groupby('Year')[['Cases', 'Deaths']].mean().reset_index()
influenza['Entity'] = 'India'
influenza['Code'] = 'IND'
influenza['Disease'] = 'Influenza'
influenza.to_csv('Yearly/yearly_influenza.csv')
influenza
```

```python
unsafe_sanitation = pd.read_csv('death-rate-from-unsafe-sanitation-gbd.csv')
unsafe_sanitation = unsafe_sanitation[unsafe_sanitation['Entity'] == 'India']
unsafe_sanitation.reset_index(inplace=True)
unsafe_sanitation.drop(columns=['index'], inplace=True)
unsafe_sanitation['Disease'] = 'Unsafe-Sanitation'
unsafe_sanitation.rename(columns={'Deaths that are from all causes attributed to unsafe sanitation per 100,000 people, in both sexes aged all ages': 'Deaths'}, inplace=True)
unsafe_sanitation['Cases'] = np.nan
unsafe_sanitation = unsafe_sanitation[['Year', 'Cases', 'Deaths', 'Entity', 'Code', 'Disease']]
unsafe_sanitation.to_csv('Yearly/yearly_unsafe-sanitation.csv')
unsafe_sanitation
```

```python
typhoid = pd.read_csv('incidence-of-invasive-non-typhoidal-salmonella-age-groups.csv')
typhoid = typhoid[typhoid['Entity'] == 'India']
typhoid.reset_index(inplace=True)
typhoid.drop(columns=['index'], inplace=True)
typhoid_sum = typhoid.iloc[:, 4:].sum(axis=1)
typhoid = pd.concat([typhoid.iloc[:, :3], typhoid_sum], axis=1)
typhoid['Code'] = 'IND'
typhoid['Disease'] = 'Typhoid'
typhoid['Deaths'] = np.nan
typhoid.rename(columns={0: 'Cases'}, inplace=True)
typhoid['Cases'] = typhoid['Cases']*100000
typhoid = typhoid[['Year', 'Cases', 'Deaths', 'Entity', 'Code', 'Disease']]
typhoid.to_csv('Yearly/yearly_typhoid.csv')
typhoid
```
